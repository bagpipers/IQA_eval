```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
csv_file_path = "./csv/sdxl-turbo_clip.csv"

# 
image_directory = "./generate_image/sdxl-turbo"

# 
data = pd.read_csv(csv_file_path)

# 
required_columns = ['Filename', 'Step', 'Prompt', 'Model', 'Width', 'Height', 'clip']
if not all(col in data.columns for col in required_columns):
    raise ValueError("CSV")

# 
if not pd.api.types.is_numeric_dtype(data['clip']):
    data['clip'] = pd.to_numeric(data['clip'], errors='coerce')
    
# NaN
data = data.dropna(subset=['clip'])

# 
widths = [512, 768, 1024]

# 
plt.figure()

for width in widths:
    # 
    filtered_data = data[data['Width'] == width]

    # clip
    average_clip = filtered_data.groupby('Step')['clip'].mean()

    # 
    plt.plot(average_clip, marker='o', label=f'Width={width}')

# 
plt.title('Average clip vs Step by Width')
plt.xlabel('Step')
plt.ylabel('Average clip')
plt.legend()
plt.grid()
plt.tight_layout()

# 
plt.savefig('clip_vs_step_by_width.png')
plt.show()
```
