```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
csv_file = "./csv/sdxl-turbo_with_overall_score.csv"
data = pd.read_csv(csv_file)

# 
def plot_overall_score(data, widths):
    plt.figure(figsize=(10, 6))

    for width in widths:
        # Width
        filtered_data = data[data['Width'] == width]

        # Stepoverall_score
        avg_scores = filtered_data.groupby('Step')['overall_score'].mean()

        # 
        plt.plot(avg_scores.index, avg_scores.values, label=f'Width={width}')

    plt.title('Average Overall Score by Step')
    plt.xlabel('Step')
    plt.ylabel('Average Overall Score')
    plt.legend()
    plt.grid()
    plt.show()

# Width
widths = [512, 768, 1024]

# 
plot_overall_score(data, widths)
```
