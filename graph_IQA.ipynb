```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
csv_file = './csv/sdxl-turbo_merged.csv'
data = pd.read_csv(csv_file)

# 
columns_to_use = ['Step', 'Width', 'arniqa', 'clipiqa', 'cnniqa', 'metaiqa']
data = data[columns_to_use]

# Width
widths = [512, 768, 1024]
metrics = ['arniqa', 'clipiqa', 'cnniqa', 'metaiqa']

# 
fig, axes = plt.subplots(1, 4, figsize=(20, 5), sharey=True)

for i, metric in enumerate(metrics):
    ax = axes[i]
    
    for width in widths:
        filtered_data = data[data['Width'] == width]
        grouped_data = filtered_data.groupby('Step')[metric].mean()
        
        ax.plot(grouped_data.index, grouped_data.values, label=f'Width {width}')
    
    ax.set_title(f'{metric.upper()} Comparison')
    ax.set_xlabel('Step')
    ax.set_ylabel('Average Value')
    ax.legend()

plt.tight_layout()
plt.show()
```
